```python
#now log tranformation to reduce the skewness in the target variable
dummy["normalized_used_price"]=np.log(dummy["normalized_used_price"])
dummy
```

```python
#now import statsmodel to bulid model

import statsmodels.api as sm 
from sklearn.model_selection import train_test_split
```

```python
x=dummy.drop("normalized_used_price",axis=1)
y=dummy["normalized_used_price"]
```

```python
x_c=sm.add_constant(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_c,y,test_size=0.3,random_state=42)
```

```python
if x_train.isna().sum().sum() == 0:
    model = sm.OLS(y_train, x_train.astype(float)).fit()  # Now should work
else:
    print("Missing values still exist! Check your preprocessing.")
```

```python
print(x_train.isna().sum())  # Check for NaN values
```

```python
x_train.fillna(x_train.mean(), inplace=True)  # Replace NaNs with column mean
```

```python
if x_train.isna().sum().sum() == 0:
    model = sm.OLS(y_train, x_train.astype(float)).fit()  # Now should work
else:
    print("Missing values still exist! Check your preprocessing.")
```

```python
model.summary()
```

```python
y_predict=model.predict(x.test.astype(float))
```

```python
residual=y_test-y_predict
residual
```

```python
#rmse
np.sqrt((residual**2).mean())
```
